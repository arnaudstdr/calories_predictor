### 1. Import des librairies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Chargement du dataset

```python
df = pd.read_csv("../data/dataset_synthetique_calories.csv")
df.head()
```

### 3. Aperçu général

```python
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

### 4. Analyse des variables
- Types de variables
- distributions

```python
df['intensite'].value_counts().plot(kind='bar')
sns.histplot(df['calories_brulees'], kde=True)
```

### 5. Corrélations

```python
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

### 6. Relations clés (calories vs autres variables)

```python
sns.pairplot(df, hue='intensite')
```

### 7. Analyse par type d'intensité

```python
sns.boxplot(data=df,x='intensite', y='calories_brulees')
```
